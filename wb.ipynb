```python
# Ссылка? Условно в реальном элемента чанка должна быть ссылка, которая потом кидается в чат
# Сейчас у нас реальных ссылок нет 
# Qdrant решает проблему передачи ссылок в чат
```

```python
# Делаем чанки

chunks = []

for i in tqdm(range(len(dataset))):
    name = dataset['name'][i]
    category = dataset['category'][i]
    keyword_category = dataset['keyword'][i]
    brand = dataset['brand'][i]
    color = dataset['color'][i]
    size = dataset['size'][i]
    description = dataset['description'][i]

    chunks.append(
f'''Название: {name}
Категория: {category}
Подкатегория: {keyword_category}
Бренд: {brand}
Цвет: {color}
Размер: {size}

Описание: {description}'''
    )
```

```python
import json

with open("chunks.json", "w", encoding="utf-8") as f:
    json.dump(chunks, f, ensure_ascii=False, indent=2)
```
